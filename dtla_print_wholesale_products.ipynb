```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

category = []
prices = []

for i in range(1, 50):
    url = "https://www.dtlaprint.com/products/page/"+str(i)+"/"
    response = requests.get(url)
    products_web_page = response.text
    soup = BeautifulSoup(products_web_page, "html.parser")


    product_details = soup.select(selector="div .post_categories")
    for item in product_details:
        texts = item.text.strip().replace("\n", "")
        category.append(texts)



    product_prices = soup.select(selector="div .price")
    for price in product_prices:
        price_text = price.text
        prices.append(price_text)


# print(category)
# print(prices)



products = pd.DataFrame({

                        "product name": category,
                        "product price": prices
})
```
